```python
def retrieve_list(df, state_list, dict_name):
 counter = 1
 for state in state_list:
  df_nar = df[df["state_po"]==state]
  df_nar2 = df_nar[(df_nar['year']==2016)&(df_nar['party_detailed']=='DEMOCRAT')].reset_index()
  demo2016 = df_nar2['proportion_votes'].iloc[0]
  df_nar3 = df_nar[(df_nar['year']==2020)&(df_nar['party_detailed']=='DEMOCRAT')].reset_index()
  demo2020 = df_nar3['proportion_votes'].iloc[0]
  df_nar4 = df_nar[(df_nar['year']==2016)&(df_nar['party_detailed']=='REPUBLICAN')].reset_index()
  rep2016 = df_nar4['proportion_votes'].iloc[0]
  df_nar5 = df_nar[(df_nar['year']==2020)&(df_nar['party_detailed']=='REPUBLICAN')].reset_index()
  rep2020 = df_nar5['proportion_votes'].iloc[0]
  indexy = counter
  dict_name[indexy] = [state, demo2016, demo2020, rep2016, rep2020]
  counter+=1
 return dict_name
```

```python
state_voting = {} 
retrieve_list(voting2, state_list, state_voting)
```

```python
display(state_voting)
```

```python
state_voting_df = pd.DataFrame(state_voting)
```

```python
state_voting_dfT = state_voting_df.transpose()
state_voting_dfT.columns = ["state", "demo2016", "demo2020", "rep2016", "rep2020"]
display(state_voting_dfT)
```
